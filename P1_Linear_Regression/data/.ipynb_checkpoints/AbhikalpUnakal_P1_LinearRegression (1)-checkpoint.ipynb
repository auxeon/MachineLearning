### Linear Regression - Using The Normal Equation

* Our Goal is to find weights $\vec{\omega}```$ that model the data as closely as possible. 
* The model $h(\vec{X})$ is a weighted linear combination of these weights $\omega_i \in \vec{\omega}$ and their respective features $x_i \in \vec{X}$
    * $h(\vec{X}) = \vec{\omega}^t \vec{X}$
    * $h(x) = \omega_0 + \omega_1x_1 + \omega_2x_2 ... \omega_nx_n$

````python
df_train.head()
````

````python
def train(df,num_features):
    column = np.full(df.shape[0],1.0)
    mat_xy = df.to_numpy()
    mat_xy = np.insert(mat_xy,0,column,axis=1)  #accounting for w0
    X = mat_xy[:,0:num_features+1]
    print(X)
    Y = mat_xy[:,num_features+1:num_features+2]
    W = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
    print(W)
    return W

def predict(df,w):
    column = np.full(df.shape[0],1.0)
    mat_xy = df.to_numpy()
    mat_xy = np.insert(mat_xy,0,column,axis=1)  #accounting for w0
    return mat_xy[:,0:len(w)].dot(w)

def plot(df,feature):
    plt.plot(df[feature],df['PRICE'],'.',color='xkcd:azure',label='REAL PRICE')
    plt.plot(df[feature],df['PREDICTED PRICE'],'+',color='xkcd:orange',label='PREDICTED PRICE')
    plt.xlabel(feature)
    plt.ylabel('PRICE')
    plt.legend(loc='best')
    plt.show()
````

### Training 

````python
W = train(df_train,1)
````

````python
regr = linear_model.LinearRegression()
print(df_train.iloc[:,0:1].head())
regr.fit(df_train.iloc[:,0:1].values,df_train.iloc[:,1:2].values)
````

````python
df_test.head()
````
